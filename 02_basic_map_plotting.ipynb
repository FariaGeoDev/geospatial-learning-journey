```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Exact path to your shapefile
path = r"C:\Users\USER\Desktop\geospatial-journey\geospatial-data\ne_110m_admin_0_countries.shp"

# Load shapefile
world = gpd.read_file(path)

# Optional: Check column names to make sure 'CONTINENT' exists
print(world.columns)

# Filter by continent
africa = world[world['CONTINENT'] == 'Africa']

# Plot
africa.plot()
plt.title("Africa Map")
plt.show()
```
